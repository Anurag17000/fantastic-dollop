```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('SampleSuperstore.csv')
```

```python
import seaborn as sns
```

```python
sns.countplot(x='Segment',data=df)
```

```python
sns.countplot(x='Category',data=df)
```

```python
sns.countplot(x=df['State'],order=(df['State'].value_counts().head(20)).index)
plt.xticks(rotation=90)
```

```python
sns.countplot(x=df['Sub-Category'],order=(df['Sub-Category'].value_counts().head(20)).index)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(14,4))
plt.subplot(131)
sns.distplot(df['Sales'])
plt.subplot(132)
sns.distplot(df['Quantity'])
```

```python
plt.figure(figsize=(14,4))
plt.subplot(133)
sns.distplot(df['Discount'])
plt.subplot(132)
sns.distplot(df['Profit'])
plt.show()
```

```python
df.groupby(by=['Sub-Category'])['Profit'].sum().plot.bar(df.groupby(by=['Sub-Category'])['Profit'].sum(),figsize=(18,9))
```

```python
df.groupby(by=['Segment'])['Profit'].sum().plot.bar(df.groupby(by=['Segment'])['Profit'].sum())
```

```python
df.groupby(by=['Region'])['Profit'].sum().plot.bar(df.groupby(by=['Region'])['Profit'].sum())
```
